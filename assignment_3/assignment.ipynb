```python
import numpy as np
import matplotlib.pyplot as plt
from function import iterate_function
```

```python
#Grid of x an y values to calcuate c

x = np.linspace(-2, 2, 500)
y = np.linspace(-2, 2, 500)

X, Y = np.meshgrid(x, y)

values = np.zeros(X.shape) #empty array to store colourscale values
binary_values = np.zeros(X.shape) #empty array to store binary colour values


#loop to interate over the points in the grid

max_iter = 150

for i in range(X.shape[0]): #vertical values (y)
    for j in range(X.shape[1]): #horizontial values (x)
        
        c = X[i, j] + 1j * Y[i, j] # complex number equaiton to cacluate c for all of the values for x and y 
        
        values[i, j] = iterate_function(c) #assigns value to location 
        
        if values[i, j] == max_iter:
            binary_values[i, j] = 1
        else:
            binary_values[i, j] = 0
         
        
        


#plot of the function

fig, ax = plt.subplots(1, 2, figsize = (12, 6))

# First plot: colorscale
im0 = ax[0].imshow(values, extent = [-2, 2, -2, 2], cmap = 'plasma', origin = 'lower')
ax[0].set_title("Colourscale Plot")
fig.colorbar(im0, ax = ax[0])

# Second plot: binary
ax[1].imshow(binary_values, extent = [-2, 2, -2, 2], cmap = 'gray', origin = 'lower')
ax[1].set_title("Binary Plot")

plt.tight_layout()
plt.show()
```
