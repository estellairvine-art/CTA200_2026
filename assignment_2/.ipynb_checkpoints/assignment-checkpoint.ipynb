```python
#Function f(x)
def f(x):
    """
    Computes the value of the function f(x) = x^3-x^2-1
    
    Parameters
    ----------- 
    x : float or int
        input value.
    
    Returns
    -----------
    float 
        The value of the function at x.
    
    """
    return x**3 - x**2 - 1

#Derivaitve of f(x)
def df(x): 
    """
    Computes the derivative of the function f(x). 
    
    The derivative is f'(x) = 3x^2-2x
    
    Parameters
    -----------
    x : float or int 
        input value
    
    Returns
    -----------
    float 
        The value of the derivative of f(x) at x.
    
    """
    
    return 3 * x**2 - 2 * x

#Newton function
def newton(f, df, x0, epsilon = 1e-6, max_iter = 30):
    """ 
    This function preforms a newton interation to aproximate the root of a funciton.
    
    Parameters
    ---------- 
    f : function
        The function whose root is being aporximated.
    df: function
        Derivative of the function whose root is being aproximated.
    x0 : float or int
        Intial guess for the root.
    esplion : float, optional
              Tolernce that determines the acuraracy of the solution (defult is 1e-6).
    max_iter : int, optional 
               Maximun number of time the function will interate (defult is 30).
    
    Returns
    ---------- 
    float or None
        Aproximate root if the functions converges sucsefully. 
        None if the method does not converge or there is a division by zero
    
    """
    
    x = x0
    
    for i in range(max_iter):
        
        if df(x) == 0:
            print('Derivative is zero')
            return None
        
        ans = x - f(x) / df(x)
        
        if abs(f(ans)) < epsilon:
            print('Found root in', i + 1, 'iterations')
            
            return ans
        
        x = ans
    
    print('Iteration falied')
    return None
```

```python
#Example with x0 set to 1

x0_1 = newton(f, df, x0 = 1)
print('Answer', x0_1)
```

```python
#Example with x0 set to 10

x0_10 = newton(f, df, x0 = 10)
print('Answer', x0_10)
```

```python
#Example with x0 set to 2/3

x0_2_3 = newton(f, df, x0 = 0)
print('Answer', x0_2_3)
```

```python
#Example with epsilon set to 1e-8 and x0 set to 1

x0_eps = newton(f, df, x0 = 1, epsilon = 1e-8)
print('Answer', x0_eps)
```

With epsilon reducded to 1e-8 the function take one more iteration to find the root. By decreasing epsilon the function requiers more iterations to be closer to zero meaning this value is more accurate. 

```python
#Epsilon to small 
x0_small = newton(f, df, x0 = 1, epsilon = 1e-50)
print('Answer', x0_small)
```
